# MULTIPAGE

```python
data_nama = []
data_cicilan = []
data_harga = []
data_alamat_penjual = []
for index in range(1,8):
    page = requests.get(f'https://www.bhinneka.com/jual?cari=iphone/page/{index}/')
    soup = BeautifulSoup(page.content, 'html.parser')
    
    nama_produk = soup.find_all('div', class_= 'product-title')
    for each_nama_produk in nama_produk:
        data_nama.append(each_nama_produk.text)
    
    cicilan = soup.find_all('span', class_= 'extra')
    for each_cicilan in cicilan:
        data_cicilan.append(each_cicilan.text)
    
    harga = soup.find_all('div', class_= 'price')
    for each_harga in harga:
        data_harga.append(each_harga.text)
    
    alamat_penjual = soup.find_all('div', class_= 'merchant-info css-ejq1bh')
    for each_alamat_penjual in alamat_penjual:
        data_alamat_penjual.append(each_alamat_penjual.text)
        

```

```python
data = {
    'data_nama': data_nama,
    'data_cicilan': data_cicilan,
    'data_harga': data_harga,
    'data_alamat_penjual': data_alamat_penjual
}
    
```
